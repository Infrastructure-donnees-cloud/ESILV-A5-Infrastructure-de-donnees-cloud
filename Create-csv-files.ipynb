```python
credit_categories_list = ["School credit loan", "Real Estate loan", "Consumer loan"]
Categories = pd.DataFrame(credit_categories_list)
Categories['category_no'] = Categories.index
Categories.columns = ['category_name','category_no']
Categories
```

```python


def get_school_loan():
    mu = 125000
    sigma = 1.2*mu
    x = int(np.random.normal(mu, sigma, 1)[0])
    return  x if x > 0 else mu

def get_real_estate_loan():
    mu = 500000
    sigma = 3*mu
    x = int(np.random.normal(mu, sigma, 1)[0])
    return  x if x > 0 else mu

def get_consummer_loan():
    mu = 5000
    sigma = 4*mu
    x = int(np.random.normal(mu, sigma, 1)[0])
    return  x if x > 0 else mu

def get_loan(category_no):
    
    loan_type = Categories[Categories['category_no']==category_no]['category_name'].values[0]
    if loan_type == "School credit loan":
        return get_school_loan()
    elif loan_type == "Real Estate loan":
        return get_real_estate_loan()
    else:
        return get_consummer_loan()
```

```python
def random_date(start, end):
    delta = end - start
    random_days = randrange(delta.days)
    return start + timedelta(days=random_days)

from datetime import datetime
d1 = datetime.strptime('1/1/2020', '%m/%d/%Y')
d2 = datetime.strptime('1/1/2022', '%m/%d/%Y')
print(random_date(d1, d2))
```

```python
loans = [[
    category_no, get_loan(category_no)
] for category_no in [randint(0, len(credit_categories_list)-1) for i in range(number_loans)]]
#loans
```

```python
loans = pd.DataFrame(loans)
loans.columns = ['loan_type', 'loan_amuont']
loans
```

```python
charge_duration = [[i*12 for i in range(1,6)][randint(0, 4)] for j in range(len(loans))]
```

```python
member_no = [randint(0, len(Members)-1) for i in range(len(loans))]
```

```python
charge_dt = [random_date(d1, d2) for i in range(len(loans))]
```

```python
Charges = pd.DataFrame([loans.loan_type.values, loans.loan_amuont.values, charge_duration, charge_dt, member_no, provider_no]).T
Charges.columns = ["Charge type", "Charge amount", "Duration", "dt", "member_no", "provider_no"]
Charges["charge_no"] = Charges.index
Charges
```

**On considère que le dataset correspond à l'état des paiements des crédit en date du 1 octobre 2021.**

**Par conséquent un emprûnt est en règle si tout les "statements" dont la due_dt est précédente au 1 octobre 2021 liés à cet umprunt ont été honorés.**


```python
dt_1er_octobre_2021 = datetime(2020, 9, 1)
dt_1er_octobre_2021
```

```python
def remboursera(row):
    if random() <= 0.98:
        return True
    else:
        return False


# In[17]:


def mois_de_defaut(row):
    if row['remboursera'] == False:
        moisDeDefaut = randint(0, row['Duration']-1)
        return moisDeDefaut
    else:
        return np.NaN
```

```python
Charges['remboursera'] = Charges.apply(lambda row: remboursera(row), axis = 1)
Charges
```

```python
Charges['Mois de défaut'] = Charges.apply(lambda row: mois_de_defaut(row), axis = 1)
```

```python
taux_mensuel = 1.002
t = [[
    [
        Charges["Charge amount"].iloc[i]/Charges["Duration"].iloc[i],
        (taux_mensuel-1)*(Charges["Charge amount"].iloc[i] - (j+1)*(Charges["Charge amount"].iloc[i]/Charges["Duration"].iloc[i])),
        Charges['dt'][i] + relativedelta(months=+j+1),
        j,
        Charges['charge_no'][i],
        Charges["member_no"][i]
    ] for j in range(Charges["Duration"][i])
] for i in range(len(Charges))]




Statements = pd.DataFrame([item for sublist in t for item in sublist])
Statements.columns = ["statement_principal", "statement_interest", "due_dt", "charge_statement_no","charge_no", "member_no"]
Statements["statement_no"] = Statements.index
Statements
```

```python
t = [[
    [
        Charges["Charge amount"].iloc[i]/Charges["Duration"].iloc[i],
        (taux_mensuel-1)*(Charges["Charge amount"].iloc[i] - (j+1)*(Charges["Charge amount"].iloc[i]/Charges["Duration"].iloc[i])),
        Charges['dt'][i] + relativedelta(months=+j+1),
        j,
        Charges['charge_no'][i],
        Charges["member_no"][i]
    ]
    for j in range(Charges["Duration"][i]) if (Charges['Mois de défaut'][i] != np.NaN) |(j < Charges['Mois de défaut'][i])
] for i in range(len(Charges))]
```

```python
Payments = pd.DataFrame([item for sublist in t for item in sublist])
Payments.columns = ["payment_principal", "payment_interest","due_dt", "charge_statement_no","charge_no", "member_no"]
Payments["payment_no"] = Payments.index
Payments
```

```python
Payments["payment_amt"] = Payments["payment_principal"] + Payments["payment_interest"]
```

```python
Payments = Payments[Payments["due_dt"] < dt_1er_octobre_2021]
Payments
```
